```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Basic Plot

```python
# Defininig the X, Y Values.
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Adding X,Y values to the plot.
plt.plot(x,y)

# Defining the X,Y points on plot.
plt.axis([1,5,1,25])

# Final Part.
plt.title("Basic Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

# Multiple Plot

```python
x = np.linspace(1,5,5)
plt.plot(x, x, label = "linear", color="green")
plt.plot(x, x**2, label = "quadratic", color="yellow")
plt.plot(x, x**3, label = "cubic", color="red")

plt.xlabel("X values")
plt.ylabel("Y values")

plt.title("Multi Plot")
plt.legend()

plt.show()
```

# Splitted Plot

```python
x = np.linspace(1,10,10)
fig,axs =  plt.subplots(3)

axs[0].plot(x, x, color="red")
axs[0].set_title("Linear")

axs[1].plot(x, x**2, color="green")
axs[1].set_title("Quadratic")

axs[2].plot(x, x**3, color="yellow")
axs[2].set_title("Cubic")

plt.tight_layout()

plt.show() 
```

# Splitted Plot v2

```python
x = np.linspace(0,10,11)
fig,axs =  plt.subplots(2,2)
fig.suptitle("Splitted Plot")

axs[0,0].plot(x, x, color="red")
axs[0,1].plot(x, x**2, color="blue")
axs[1,0].plot(x, x**3, color="green")
axs[1,1].plot(x, x**4, color="yellow")

plt.show() 
```

# Mathematical Plot

```python
x = np.linspace(-25,24,25)
y = x ** 3
z = x ** 2

fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

axes[0].plot(x,y)
axes[0].set_title("f(x)=x^3")
axes[1].plot(x,z)
axes[1].set_title("f(x)=x^2")
plt.tight_layout()

plt.show()
```

# Pandas Plot

```python
# Gathering the data.

"""
data = pd.read_csv("dataset.csv")
df = pd.DataFrame(data)
"""

# or

dictionary = {"Name":["Lewis","Max","Sebastian","Fernando","Schumacher"],
              "WDC":[7,1,4,2,7]}

data = pd.DataFrame(dictionary)

# Preparing the data.

names = []
values = []

for i in data.iloc:
    names.append(i[0])
    values.append(i[1])

# Creating the plot.

plt.bar(names,values,label="F1 TOP TIER CHAMPIONS",width=.5)

plt.legend()
plt.xlabel("Drivers")
plt.ylabel("Championships")
plt.title("Formula 1 Championship")

plt.show()
```

# Pie Plot

```python
# Shadow & Explode for Much better visualization. The f%%f shows the ratio.

goal_types = 'National Teams','Young Teams','Club Teams'

goals = [102,18,660]
colors = ['y','r','b']

plt.pie(goals,labels=goal_types,colors=colors, shadow=True, explode=(0.05,0.05,0.05))
plt.show()
```

# Stack Plot

```python
years = [1,2,3]
lebron = [20.9, 27.2, 31.4]
durant = [20.3, 25.3, 30.1]
steph = [17.5, 18.6, 14.7]

plt.plot([],[],color="y",label="LeBron James")
plt.plot([],[],color="r",label="Kevin Durant")
plt.plot([],[],color="b",label="Stephen Curry")

plt.stackplot(years,lebron,durant,steph, colors=["y","r","b"])
plt.title("SCORERS")
plt.xlabel("Year")
plt.ylabel("PPG")
plt.legend()
plt.show()
```
